```python
from torchvision.datasets import ImageFolder
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
dataset = ImageFolder(
    "../data/raw/PlantVillage"
)
```

```python
from collections import Counter

targets = dataset.targets
class_counts = Counter(targets)

for idx, count in class_counts.items():
    print(
        f"{dataset.classes[idx]} : {count}"
    )
```

```python
## Plotting barplot

plot_data = {dataset.classes[idx] : count for idx, count in class_counts.items()}

classes = list(plot_data.keys())
counts = list(plot_data.values())

plt.figure(figsize=(10,6))
sns.barplot(x=counts, hue=classes, legend=False, palette='magma')
plt.title('PlantVillage Class Distribution')
plt.xlabel('Count')
plt.show()
```

```python
import random
fig, axes = plt.subplots(3,3,figsize=(10,10))
for ax in axes.flat:
    idx = random.randint(0,len(dataset)-1)
    image, label = dataset[idx]
    ax.imshow(image)
    ax.set_title(dataset.classes[label])
    ax.axis("off")

plt.tight_layout()
plt.show()
```
